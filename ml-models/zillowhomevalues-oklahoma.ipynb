```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
from io import StringIO
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# I had to create a workaround to use my CSV files uploaded to Google Drive
url = 'https://drive.google.com/file/d/1beJ4WIuq1oFtW78v4zeszPCuX1oyDxJf/view?usp=sharing'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
print(df.head())

# file_path = '../resources/zillowhomevalues_tulsacounty.csv'
# df = pd.read_csv(file_path)
# print(df.head())
```

```python
print(df.head())
print(df.columns)
```

```python
# Strip whitespaces from column names
df.columns = [col.strip() if isinstance(col, str) else col for col in df.columns]

# Reshape the data
date_columns = df.columns[8:]
reshaped_data = df.melt(id_vars=['RegionName'], value_vars=date_columns, var_name='Date', value_name='AveragePrice')
reshaped_data['Date'] = pd.to_datetime(reshaped_data['Date'])
```

```python
# Calculate the number of months since the starting date (1/31/2000)
reshaped_data['MonthsSinceStart'] = (reshaped_data['Date'] - pd.to_datetime('2000-01-31')).dt.days // 30
```

```python
# Fill missing values with the mean of the column
reshaped_data['AveragePrice'].fillna(reshaped_data['AveragePrice'].mean(), inplace=True)
```

```python
# Drop rows with NaN values in the 'AveragePrice' column
reshaped_data.dropna(subset=['AveragePrice'], inplace=True)
```

```python
# Split the data into training and testing sets
X = reshaped_data[['MonthsSinceStart']]
y = reshaped_data['AveragePrice']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Fit the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Plot the training data and the regression line
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.scatter(X_test, y_test, color='red', label='Testing data')
plt.plot(X, model.predict(X), color='black', label='Linear regression')
plt.xlabel('Months Since Start')
plt.ylabel('Average Home Price')
plt.legend()

# Save the plot to a PNG file
plt.savefig('../images/zillowdata-tulsaCo_regressionplot.png', dpi=300)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Calculate the evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")
print(f"R-squared: {r2:.2f}")
```
