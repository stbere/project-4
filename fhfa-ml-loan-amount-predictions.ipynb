```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Load the CSV file into a Pandas DataFrame
df = pd.read_csv('../resources/fhfagov-2021.csv')
```

```python
# Choose the target variable and independent variables
target_var = 'NoteAmount'
ind_vars = ['TotalMonthlyIncomeAmount', 'LTVRatioPercent']
```

```python
# Split the data into training and testing sets
train_size = 0.8
train_idx = int(train_size * len(df))
train_data = df.iloc[:train_idx]
test_data = df.iloc[train_idx:]
```

```python
# Fit a linear regression model to the training data
model = LinearRegression()
model.fit(train_data[ind_vars], train_data[target_var])
```

```python
# Ignore the warning message resulting from making a prediction on the test data
pd.options.mode.chained_assignment = None
```

```python
# Make predictions on the test data
test_data['predictions'] = model.predict(test_data[ind_vars])
```

```python
# Plot the results
plt.scatter(test_data[target_var], test_data['predictions'])
plt.xlabel('Actual Loan Amount')
plt.ylabel('Predicted Loan Amount')
plt.title('Linear Regression Model')
plt.show()
```

```python
# Save to images folder
fig = plt.figure()
fig.savefig('../images/fhfa-ml_linear.regression.png')
```
